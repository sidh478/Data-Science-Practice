```python
#import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#to display all columns for dataframe
pd.set_option('display.max_columns',None)
```

```python
#reading data set
dataset=pd.read_excel(r'C:\Users\harsh\Downloads\krish naik\EDA\Flight Prediction\Data_Train.xlsx')
dataset.head()
```

```python
dataset.info()
```

```python
#Date_of_Journey is spliting into date,month&year
#splitting can be done in 2 ways
dataset['Date']=dataset['Date_of_Journey'].str.split('/').str[0]
dataset['Month']=dataset['Date_of_Journey'].str.split('/').str[1]
dataset['Year']=dataset['Date_of_Journey'].str.split('/').str[2]
```

#### or

```python
dataset['Date']=dataset['Date_of_Journey'].apply(lambda x:x.split('/')[0])
dataset['Month']=dataset['Date_of_Journey'].apply(lambda x:x.split('/')[0])
dataset['Year']=dataset['Date_of_Journey'].apply(lambda x:x.split('/')[0])
```

```python
#Date_of_Journey hase be splitted into date,month&year but still its not integer
dataset.info()
```

```python
#converting object into integer(date,month&year)
dataset['Date']=dataset['Date'].astype(int)
dataset['Month']=dataset['Month'].astype(int)
dataset['Year']=dataset['Year'].astype(int)
```

```python
#sucessfully converted object to integer.
dataset.info()
```

```python
#Hence we don't need Date_of_Journey column.So,drop it.
dataset.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
dataset.head()
```

```python
#splitting arrival time into hour,min,month
dataset['Arrival_Time']=dataset['Arrival_Time'].str.split(' ').str[0]
#splitting by ':' and taking hour index 0
dataset['Arrival_Hour']=dataset['Arrival_Time'].str.split(':').str[0]
#splitting by ':' and taking min index 1
dataset['Arrival_Min']=dataset['Arrival_Time'].str.split(':').str[1]
```

```python
#converting object into integer
dataset['Arrival_Hour']=dataset['Arrival_Hour'].astype(int)
dataset['Arrival_Min']=dataset['Arrival_Min'].astype(int)
```

```python
#splitting arrival time into hour,min,month
dataset['Dep_Time']=dataset['Dep_Time'].str.split(' ').str[0]
#splitting by ':' and taking hour index 0
dataset['Dep_Hour']=dataset['Dep_Time'].str.split(':').str[0]
#splitting by ':' and taking min index 1
dataset['Dep_Min']=dataset['Dep_Time'].str.split(':').str[1]
```

```python
#converting object into integer
dataset['Dep_Hour']=dataset['Dep_Hour'].astype(int)
dataset['Dep_Min']=dataset['Dep_Min'].astype(int)
```

```python
#dropping 'Dep_Time','Arrival_Time' as we splitted into hour and min
dataset.drop(['Dep_Time','Arrival_Time'],axis=1,inplace=True)
```
