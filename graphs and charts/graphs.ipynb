```python
df=pd.read_csv('https://raw.githubusercontent.com/venky14/iris-dataset/master/iris-species/Iris.csv')
df
```

```python
#id column is droped as its disturbing graph visualisation
df=df.drop('Id',axis=1)
df
```

```python
df.plot(figsize=(20,10))
```

```python
df.iloc[5:11].plot(kind='bar',figsize=(20,10))
```

```python
df.iloc[[5]].plot(kind='barh',figsize=(8,4))
```

```python
#x axis represent data range
#y=axis represent data frequency
df['SepalLengthCm'].plot(kind='hist',orientation='horizontal')
```

```python
#color change bar color
#alpha sets density of colour
df.hist(figsize=(20,10),color='g',alpha=.5)
```

```python
#can use any below
df.plot(kind='box',color='r')
df.plot(kind='box',color={'boxes':'g','whiskers':'r'})
df.plot(kind='box',color={'boxes':'g','whiskers':'r'},vert=False)
```

```python
df.describe()
```

```python
df.plot(kind='area',alpha=.5,stacked=False)
```

```python
df.plot(kind='area',alpha=.5,stacked=True)
```

```python
df.plot.scatter(x='SepalLengthCm',y='PetalLengthCm')
```

```python
#c='PetalWidthCm' indicates colour code based on the value of data
#s=60 changesthe size of the circle
df.plot.scatter(x='SepalLengthCm',y='PetalLengthCm',c='PetalWidthCm',s=60)
```

```python
df.plot.scatter(x='SepalLengthCm',y='PetalLengthCm',c='PetalWidthCm',s=df['PetalWidthCm']*500)
```

```python
#gridsize=10 indicates size of grid
#capital C should be used
df.plot.hexbin(x='SepalLengthCm',y='PetalLengthCm',gridsize=10,C='SepalWidthCm')
```
